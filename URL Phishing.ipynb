```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import svm
```

```python
# From https://www.kaggle.com/datasets/sunnykusawa/phishing-urls?resource=download
url_df = pd.read_csv("phishing_url_dataset.csv")

# Here, we're sampling the data so that we can narrow down our focus on
# the graphs to 2000 data points
sample_df = url_df.sample(2000)

# Showing the sample dataset
sample_df
```

```python
# Now, we will group the data by valid URL and isHTTPs columns for
# the first analysis
grouped_data = sample_df.groupby(['valid_url', 'isHttps'])

# Counting the number of each occurence
group_size = sample_df.groupby(['valid_url', 'isHttps']).size()

# INVALID URL PI CHART:

# Labeling the sections of the pi chart to be HTTP or HTTPS
valid_url_labels = ['HTTP', 'HTTPS']

# Getting the sizes/amounts of the invalid URLS
valid_url_sizes = group_size.loc[0]

# Making the new figure and setting the size to be 12,6. We will be
# consistent for all of our illustrations in this analysis, so all
# of the figures will have a size of 12,6.
plt.figure(figsize=(12, 6))

# Creating a subplot with one row and two columns. Putting the invalid
# URLs on the right hand side. Again, we will be consistent throught the
# analysis and all of the "invalid URL" findings will be on the right.
plt.subplot(1, 2, 2)

# Making the pi chart. Labeling invalid URL and setting the percentage
# to only have one decimal point. We also set the start angle to be
# 180, which does not change the analysis, it just causes the pi
# graph to start at the left hand side and rotate around clockwise.
plt.pie(valid_url_sizes, labels=valid_url_labels, autopct='%1.1f%%',
        startangle=180, colors=['#cc0000', 'green'])

# Giving the invalid pi chart a title
plt.title('Invalid URLs')

# VALID URL PI CHART:

# Labeling the sections of the pi chart to be HTTP or HTTPS
invalid_url_labels = ['HTTP', 'HTTPS']

# Getting the sizes/amounts of the valid URLS
invalid_url_sizes = group_size.loc[1]

# Creating a subplot with one row and two columns. Putting the valid
# URLs on the left hand side. Again, we will be consistent throught the
# analysis and all of the "valid URL" findings will be on the left.
plt.subplot(1, 2, 1)

# Making the pi chart. Labeling valid URL and setting the percentage
# to only have one decimal point. We also set the start angle to be
# 180, which does not change the analysis, it just causes the pi
# graph to start at the left hand side and rotate around clockwise.
plt.pie(invalid_url_sizes, labels=invalid_url_labels,
        autopct='%1.1f%%', startangle=180, colors=['#cc0000', 'green'])

# Giving the valid pi chart a title
plt.title('Valid URLs')

# Plotting the pi charts
plt.show()
```

```python
# Please be patient with this illustration loading. It takes about
# 10-15 seconds due to the number of datapoints. We have tried using a
# lower sample number to take less time, but then we found it would not
# show as many data points as we would have liked

# First, we are filtering out sample_df to extract the rows where the
# URL is "valid", which is when the int value is equal to one.
valid_urls = sample_df[sample_df['valid_url'] == 1]

# Then, we will filtering out sample_df again to extract the rows where the
# URL is "invalid", which is when the int value is equal to zero.
invalid_urls = sample_df[sample_df['valid_url'] == 0]

# Counting the number of valid URLs from the new extracted rows
valid_count = len(valid_urls)

# Counting the number of invalid URLs from the new extracted rows
invalid_count = len(invalid_urls)

# Making the new figure and setting the size to be 12,6.
plt.figure(figsize=(12, 6))

# INVALID URL BAR CHART:

# Creating a subplot with one row and two columns. Putting the invalid
# URLs on the right hand side. Again, we will be consistent throught the
# analysis and all of the "invalid URL" findings will be on the right.
plt.subplot(1, 2, 2)

# Creating a bar chart for invalid URLs showing URL length
plt.bar(range(invalid_count), invalid_urls['url_length'],
        label='URL Length', color='#7f0000')

# Adding onto the invalid URL bar chart to showing URL path length
plt.bar(range(invalid_count), invalid_urls['path_length'],
        bottom=invalid_urls['url_length'], label='Path Length', color='red')

# Adding an x-axis label for number of invalid URLs
plt.xlabel('Number of Invalid URLs')

# Adding a y-axis label for length of URL or path length
plt.ylabel('Length')

# Adding a title to the invalid URL bar chart
plt.title('Invalid URLs')

# Adding a legend to show the color differences between the showed
# path lengths and URL lengths
plt.legend()

# VALID URL BAR CHART:

# Creating a subplot with one row and two columns. Putting the valid
# URLs on the left hand side. Again, we will be consistent throught the
# analysis and all of the "valid URL" findings will be on the left.
plt.subplot(1, 2, 1)

# Creating a bar chart for valid URLs showing URL length
plt.bar(range(valid_count), valid_urls['url_length'],
        label='URL Length', color='green')

# Adding onto the valid URL bar chart to showing URL path length
plt.bar(range(valid_count), valid_urls['path_length'],
        bottom=valid_urls['url_length'], label='Path Length', color='#00b200')

# Adding an x-axis label for number of valid URLs
plt.xlabel('Number of Valid URLs')

# Adding a y-axis label for length of URL or path length
plt.ylabel('Length')

# Adding a title to the valid URL bar chart
plt.title('Valid URLs')

# Adding a legend to show the color differences between the showed
# path lengths and URL lengths
plt.legend()

# Plotting the bar charts
plt.show()
```

```python
# First, we are filtering out sample_df to extract the rows where the
# URL is "valid", which is when the int value is equal to one.
valid_urls = sample_df[sample_df['valid_url'] == 1]

# Then, we will filtering out sample_df again to extract the rows where the
# URL is "invalid", which is when the int value is equal to zero.
invalid_urls = sample_df[sample_df['valid_url'] == 0]

# This is the most important section of the characteristic analysis. Here,
# we decided to make a python dictionary of all of the different
# characteristics that make up a URL. These different characterics come
# from the dataset that we have chosen. We are using key/value pairs of
# the column name of the characteric from the dataset and then what
# its corresponding title will be for each valid/invalid URL graph.
characteristic = {
    'nb_dots': 'Number of Dots',
    'nb_hyphens': 'Number of Hyphens',
    'nb_underscore': 'Number of Underscores',
    'nb_and': 'Number of Ands',
    'nb_or': 'Number of Ors',
    'nb_www': 'Number of www\'s',
    'nb_com': 'Number of .com\'s'
}

# This specific function is going to handle plotting the different
# characteristics of a URL from our data in the dataset. Each time
# this function is run, it plots the invalid and valid graphs for
# each characteristic. The parameters include the "characteristic" and
# the corresponding title/label for each characteric. This was mentioned
# above in the dictionary section.
def plot_combined_histogram(valid_data, invalid_data, characteristic, title):
    # Making the new figure and setting the size to be 12,6.
    plt.figure(figsize=(12, 6))

    # Adding a title to each characteric section. Dosnt really add
    # any value, but its helping to separate out the characterics
    plt.suptitle('URL Phishing Characteristics', fontsize=16)

    # VALID URL HISTOGRAM FOR EACH CHARACTERISTIC:

    # Creating a subplot with one row and two columns. Putting the valid
    # URLs on the left hand side. Again, we will be consistent throught the
    # analysis and all of the "valid URL" findings will be on the left.
    plt.subplot(1, 2, 1)

    # Plotting histogram for valid URLs. Here, the "kde=true" is a very
    # helpful addition because it allows us to see a line graph on top
    # of the histogram to see the average changes in the data
    sns.histplot(data=valid_data, x=characteristic,
                 color='green', bins=20, kde=True)

    # Counting the number of times that the specific characteric
    # was found in the URL
    plt.xlabel('Count')

    # Adding an y-axis label for count of number of characterics
    # found in the dataset given if the URL was determined to be valid
    plt.ylabel('Frequency')

    # Adding a title to the valid URL histogram. Using the parameter
    # to get the label from the characteristic dictionary created above
    plt.title(f'Valid URLs - {title}')

    # Counting the maximum value for the frequency that a given count
    # of a characteric appears in the dataset for valid data. This is
    # important because we want to ensure that the heights of the
    # y-axis are the same for both valid and invalid graphs. This
    # will ensure the analysis of the graphs are not skewed, biased,
    # or wrong.
    valid_max_count = valid_data[characteristic].value_counts().max()

    # Counting the maximum value for the frequency that a given count
    # of a characteric appears in the dataset for valid data. This is
    # important because we want to ensure that the heights of the
    # y-axis are the same for both valid and invalid graphs. This
    # will ensure the analysis of the graphs are not skewed, biased,
    # or wrong.
    invalid_max_count = invalid_data[characteristic].value_counts().max()

    # Getting the maximum height of either the valid or invalid
    # graph to figure out what to set the height of the y-axis to be.
    combined_max_count = max(valid_max_count, invalid_max_count)

    # Setting the y-axis height of the valid URL graph based on the
    # maximum height found between both valid or invalid data. Here,
    # we also multiplied the maximum height by 1.1 to allow for extra
    # space at the top of the graph. Before this, the top of the
    # histogram was touching the top of the graph and it did not
    # look good.
    plt.ylim(0, combined_max_count * 1.1)

    # INVALID URL HISTOGRAM FOR EACH CHARACTERISTIC:

    # Creating a subplot with one row and two columns. Putting the invalid
    # URLs on the right hand side. Again, we will be consistent throught the
    # analysis and all of the "invalid URL" findings will be on the right.
    plt.subplot(1, 2, 2)

    # Plotting histogram for invalid URLs. Here, the "kde=true" is a very
    # helpful addition because it allows us to see a line graph on top
    # of the histogram to see the average changes in the data
    sns.histplot(data=invalid_data, x=characteristic,
                 color='#cc0000', bins=20, kde=True)

    # Counting the number of times that the specific characteric
    # was found in the URL
    plt.xlabel('Count')

    # Adding an y-axis label for count of number of characterics
    # found in the dataset given if the URL was determined to be invalid
    plt.ylabel('Frequency')

    # Adding a title to the invalid URL histogram. Using the parameter
    # to get the label from the characteristic dictionary created above
    plt.title(f'Invalid URLs - {title}')

    # Setting the y-axis height of the valid URL graph based on the
    # maximum height found between both valid or invalid data. Here,
    # we also multiplied the maximum height by 1.1 to allow for extra
    # space at the top of the graph. Before this, the top of the
    # histogram was touching the top of the graph and it did not
    # look good.
    plt.ylim(0, combined_max_count * 1.1)

    # Plotting the valid and invalid histograms for each characteristic
    plt.show()

# Here's where the function showed below is actually being called for
# each characteric in the dictionary
for characteristic, label in characteristic.items():
    # Calling the function explained above.
    plot_combined_histogram(valid_urls, invalid_urls, characteristic, label)
```

```python
# Now that we have showed our findings of the invalid and valid URL
# data from our chosen dataset, we will use an LLM training model
# of train_test_split to see if the LLM can predict if a URL is
# valid or invalid given a set of inputted URLS

# This is an important step that we found while doing this project. We
# got very stuck on trying to figure out how to add all of these
# different characterics into the model. train_test_split only allows
# for a certian number of parameters and most of our examples and homework
# from class did not have this many features. After many failed attempts,
# we found that this way worked. We put all of the different characterics
# into a single array
features = url_df[['isHttps', 'nb_dots', 'nb_hyphens', 'nb_underscore',
                   'nb_and', 'nb_or', 'nb_www', 'nb_com']]

# Adding our label of invalid URL or not for the train_test_split model
label = url_df['valid_url']

# Iterate over a range of test sizes from 9% to 90% with a step size
# of 10%. This is something that we did in class and were interested to
# see what the different accuracys would look like for different
# percentages of training versus testing in our project here too
# Citation: This code came from the lectures in class
for percentage in range(9, 100, 10):
    contents_train, contents_test, label_train, label_test = train_test_split(features, label, test_size=percentage / 100)
    model = svm.SVC()
    model.fit(contents_train, label_train)
    print("Test size percentage: {}% and Accuracy: {}".
          format(percentage, model.score(contents_test, label_test)))

# This section splits the data into training and testing splits using a
# test size of 20%. We decided on 20% because the for-loop section above (which
# tested all of the different accuracies for the step sizes of 10%) had the
# highest accuracy for 20% at 0.88 when we ran it.
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.2)

# Train the model on the data
model = svm.SVC()
model.fit(X_train, y_train)

# Getting the accuracy of the most recently trained model. This model will
# be the one that we use to compare our hand-tested URLs.
accuracy = model.score(X_test, y_test)

# Printing out the accuracy of the most recently trained model.
print('\n')
print("Accuracy for 20% test size used to train prediction model below:", accuracy)

# List of hand-tested URLs. Citation: These were provided in the lectures
hand_test = [
    "https://www.section.io/engineering-education/",
    "https://www.youtube.com/",
    "https://www.traversymedia.com/",
    "https://www.kleinehundezuhause.com/",
    "http://ttps//www.mecymiafinance.com",
    "https://www.atlanticoceanicoilandgas.com/",
    "http://www.buyfakebillsonlinee.blogspot.com/",
    "http://www.unitedairlineslogistics.com/",
    "http://www.stonehousedelivery.com/",
    "http://www.silkroadmeds-onlinepharmacy.com/"
]

# This is where we did the logic for counting the number of characterics
# in each of the hand-tested URLs. We added each of these counts to a
# dataframe so we could use them in the model predictions.
test_features = pd.DataFrame({
    'isHttps': [url.startswith('https://') for url in hand_test],
    'nb_dots': [url.count('.') for url in hand_test],
    'nb_hyphens': [url.count('-') for url in hand_test],
    'nb_underscore': [url.count('_') for url in hand_test],
    'nb_and': [url.count('&') for url in hand_test],
    'nb_or': [url.count('|') for url in hand_test],
    'nb_www': [url.count('www') for url in hand_test],
    'nb_com': [url.count('.com') for url in hand_test]
})

# Predicting whether the URLs are invalid or not based on the features
# and characterics of the hand-tested URL's
predictions = model.predict(test_features)

# Printing out the model AI predicitions based on the training, testing,
# and inputted data
print('\n')
print('Valid/Invalid URL AI Predictions:')

# For each hand-tested URL, this will print if the model predicted the
# hand-tested URL as invalid or valid
for i in range(len(hand_test)):
    # Getting an array of the hand-tested URLs
    url = hand_test[i]

    # Getting an array of the predictions returned by the model
    prediction = predictions[i]

    # Checks to see if the model returned the URL as invalid or valid
    if prediction:
        message = "Valid"
    else:
        message = "Invalid"

    # Prints out the URL and if the model predicted valid or invalid
    print(url, "->", message)
```
